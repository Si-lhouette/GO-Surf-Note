## 原数据集读取
- 原来的数据集的深度图的数值是实际深度的1000倍

```python
def alphanum_key(s):
    """ Turn a string into a list of string and number chunks.
        "z23a" -> ["z", 23, "a"]
    """
    return [int(x) if x.isdigit() else x for x in re.split('([0-9]+)', s)]

def onclick(event):
    if event.xdata is not None and event.ydata is not None:
        row_index = int(event.ydata)
        col_index = int(event.xdata)
        depth_value = depth[row_index, col_index]
        print("Depth value at ({}, {}): {}".format(row_index, col_index, depth_value))
```

```python
datesetdir = "/home/michael/Recon/go_surf_ws/src/go-surf/neural_rgbd_data"
basedir = os.path.join(datesetdir, "grey_white_room")

depth_files = [f for f in sorted(os.listdir(os.path.join(basedir, 'depth_filtered')), key=alphanum_key) if f.endswith('png')]
```

```python
frame_id = 1

depth = imageio.imread(os.path.join(basedir, 'depth_filtered', depth_files[frame_id]))

# 显示深度图像
plt.imshow(depth, cmap='gray')
plt.connect('button_press_event', onclick)
plt.axis('off')  # 关闭坐标轴
plt.colorbar()   # 添加颜色条
plt.show()
```
